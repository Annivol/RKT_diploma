```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df = pd.read_csv('продажи_2023_со_всеми_характеристиками.csv', encoding='utf-8', sep=',', low_memory=False)
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
from datetime import datetime
df['date_time_0'] = df['date_time_0'].astype("datetime64[ns]")
df = df.copy()
df['quarter'] = df['date_time_0'].dt.quarter
```

```python
df1 = pd.get_dummies(df, columns=['quarter']).copy()
```

```python
df1
```

```python
le.fit(df1['uid_doc'])
uid_docs_le = le.transform(df1['uid_doc'])
df1 = df1.copy()
df1['uid_docs_le'] = uid_docs_le
```

```python
le.fit(df1['customer_key'])
customer_keys_le = le.transform(df1['customer_key'])
df1 = df1.copy()
df1['customer_keys_le'] = customer_keys_le
```

```python
df1
```

```python
df_nano_tocheck = df1.iloc[:,[1,12,13, 14, 143, 144, 145, 146, 147, 148]]
df_nano_toteach = df1.iloc[:,[12,13, 14, 143, 144, 145, 146, 147, 148]]
df_nano_tocheck.head(5)
```

```python
X = df_nano_toteach.iloc[:,:-1]
y = df_nano_toteach.iloc[:,-1:].to_numpy().ravel()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 50)
```

```python
clf.fit(X_train, y_train)
```

```python
print("Прогноз на тестовом наборе: {}".format(clf.predict(X_test)))
```

```python
print ("Правильность на тестовом наборе: {}".format(clf.score(X_test, y_test)))
```

```python
uid_dc_labels = clf.predict(X)
```

```python
df_nano_tocheck['clusters'] = uid_dc_labels

df_nano_tocheck.head(5)
```
