```python
from sklearn.preprocessing import StandardScaler
stdscaler = StandardScaler()
X_stdscaled = stdscaler.fit_transform(df_toteach)
```

```python
from sklearn.decomposition import PCA
pca0 = PCA(n_components = 2, whiten = True, random_state = 0)
pca0.fit_transform(X_stdscaled)
X_pca0 = pca0.transform(X_stdscaled)
plt.plot(X_pca0[:,0], X_pca0[:,1],'bo')
```

```python
from matplotlib import colormaps
plt.figure(figsize = (20,5))
plt.matshow(pca0.components_, cmap = 'viridis', fignum = 1, aspect = 'auto')
plt.yticks([0,1], ['Первая', 'Вторая'])
plt.colorbar()
```

```python
def analyzePCA(X, pca, explained_threshold=0.95):
    """Plot cumulative explaned variance of PCA"""
    pca.fit(X)
    pca_variance = pca.explained_variance_ratio_
    plt.figure(figsize=(15,7))
    plt.plot(np.cumsum(pca_variance))
    plt.hlines(explained_threshold, 0, len(pca_variance))
    plt.title("Number of components to explain {}% of variance: {}".format(
        100*explained_threshold, np.argwhere(np.cumsum(pca_variance)>0.95)[0]
    ))
    plt.xlabel('number of components')
    plt.ylabel('cumulative explained variance');
```

```python
pca = PCA()
analyzePCA(X_stdscaled, pca)
```

```python
pcadbs = PCA(n_components = 280, whiten = True, random_state = 0)
pcadbs.fit_transform(X_stdscaled)
X_pcadbs = pcadbs.transform(X_stdscaled)
```

```python
from sklearn.cluster import DBSCAN
dbscan1 = DBSCAN(eps = 10, min_samples = 20)
labels = dbscan1.fit_predict(X_pcadbs)
print("Уникальные метки: {}".format(np.unique(labels)))
print("размеры кластеров: {}".format(np.bincount(labels+1))) 
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 300, random_state = 1).fit(X_pcadbs)
```

```python
df_toteach['clusters'] = kmeans.labels_
df_tocheck['clusters'] = kmeans.labels_
df_tocheck.head(5)
```
