```python
# Пример про лица из книги. Берем выделяем компоненты сначала, потом делаем кластеризацию DBSCAN
from sklearn.decomposition import PCA
pcadbs = PCA(n_components = 200, whiten = True, random_state = 0)
pcadbs.fit_transform(X_scaled_dm0)
X_pcadbs = pcadbs.transform(X_scaled_dm0)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN()
labels = dbscan.fit_predict(X_pcadbs)
print("Уникальные метки: {}".format(np.unique(labels)))
```

```python
# прям долго считал, но чот даже много 0.85
from sklearn.metrics.cluster import silhouette_score
print("silhouette: {}".format(silhouette_score(X_pcadbs, labels)))
```

```python
# то же самое, но уменьшим количество компонент до 100
pcadbs1 = PCA(n_components = 100, whiten = True, random_state = 0)
pcadbs1.fit_transform(X_scaled_dm0)
X_pcadbs1 = pcadbs1.transform(X_scaled_dm0)
dbscan1 = DBSCAN()
labels1 = dbscan1.fit_predict(X_pcadbs1)
print("Уникальные метки: {}".format(np.unique(labels1)))
```

```python
# А теперь возьмем ненормированные данные (без scaler)
pcadbs2 = PCA(n_components = 200, whiten = True, random_state = 0)
pcadbs2.fit_transform(df_testtoteach)
X_pcadbs2 = pcadbs1.transform(df_testtoteach)
dbscan2 = DBSCAN()
labels2 = dbscan2.fit_predict(X_pcadbs2)
print("Уникальные метки: {}".format(np.unique(labels2)))
```
