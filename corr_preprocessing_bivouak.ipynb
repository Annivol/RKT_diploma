```python
def nohenc(X, column) -> pd.DataFrame:
    nohcols = []
    uniq = X[column].unique()  # это список с вариациами наборов вариантов значений разделенных слешами
    for i in range(len(uniq)):  
        coldata = uniq[i] # тут строка из одной ячейки с набором вариантов и слешами              
        act = str(coldata).split('/')  # и почему-то тут он дает ошибку что не может рассплитить float.... 
        for j in range(len(act)):
            nohcols.append(act[j])
    newcolumns = set(nohcols)    
    for i in range(X.shape[0]):
        for col in newcolumns:
          X[col] = 0 
    for i in range(X.shape[0]):
        coldata= X[column][i]
        X=X.copy()
        for col in newcolumns:
            if col in str(coldata):
                X[col][ i] = 1
    del X['']
    return X
```
