```python
import pandas as pd
```

```python
d = {'name':['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k'], 'activity':[ 'tour/city/mount/', 'tour/', 'city/', 'mount/','tour/city/','mount/city/', 'mount/tour/', 'tour/mount/', '','forest/', 'climb/tour/'] }
df = pd.DataFrame(data=d)
df
```

```python
separator = '/'
nohcols = []
for i in range(df.shape[0]):
    coldata= df.activity[i]
    coldata = coldata[:-1]
    act = coldata.split('/')
    for j in range(len(act)):
        nohcols.append(act[j])
newcolumns = set(nohcols)
newcolumns   
```

```python
for i in range(df.shape[0]):
    for col in newcolumns:
      df[col] = 0 
df
```

```python
for i in range(df.shape[0]):
    coldata= df.activity[i]
    coldata = coldata[:-1]
    for col in newcolumns:
        if col in coldata:
            df[col][ i] = 1
del df['']
df           
```
